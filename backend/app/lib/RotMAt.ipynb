```python
def getV(a, dx, dy, dz):
    ''' 
    Функция получения матрицы поворота стержня
    по заданным координатам вектора и углу поворота оси
        Параметры:
            a - угол поворота вокруг собственной оси стержня, радианы
            dx, dy, dz - координаты вектора,  dx = xj - xi, dy = yj - yi,  dz = zj - zi
    Возвращает:
        Матрица поворота стержня V
    '''

    # Вспомогательные функции для определения угла между векторами
    dotproduct = lambda v1, v2: sum((a*b) for a, b in zip(v1, v2))
    length = lambda v: math.sqrt(dotproduct(v, v))
    angle = lambda v1, v2: math.acos(dotproduct(v1, v2) / (length(v1) * length(v2)))

    # 1. Определяем угол между проекцией оси стержня на плоскость [YZ] и осью Y
    angYZ_Y = 0
    if dy == 0 and dz == 0:
        #pass
   # elif dy == 0 and dx == 0:
        angYZ_Y = math.pi / 2

    else:
        angYZ_Y = angle([0, 1, 0], [0, dy, dz])
    if dz < 0:
         angYZ_Y = - angYZ_Y

    print(math.degrees(angYZ_Y))

    # 2. Матрица первого поворота вокруг оси X
    V1 = np.array([[1, 0, 0], 
                   [0, math.cos(angYZ_Y), - math.sin(angYZ_Y)], 
                   [0, math.sin(angYZ_Y), math.cos(angYZ_Y)]])

    # 3. Определяем вектор повернутый относительно оси X
    vX = V1.T.dot([dx, dy, dz])
    
    # 4. Определяем угол между вектором vX и осью X
    angvX_X = angle([1, 0, 0], vX)

    # 5. Матрица второго поворота вокруг оси Z
    V2 = np.array([[math.cos(angvX_X), - math.sin(angvX_X), 0], 
                   [math.sin(angvX_X), math.cos(angvX_X), 0], 
                   [0, 0, 1]])

    # 6. Матрица поворота вокруг собственной оси на угол a
    V3 = np.array([[1, 0, 0], 
                   [0, math.cos(a), - math.sin(a)], 
                   [0, math.sin(a), math.cos(a)]])
    

  
    
    return V1.dot(V2.dot(V3))


def getV2(a, dx, dy, dz):
    ''' 
    Функция получения матрицы поворота стержня
    по заданным координатам вектора и углу поворота оси
        Параметры:
            a - угол поворота вокруг собственной оси стержня, радианы
            dx, dy, dz - координаты вектора,  dx = xj - xi, dy = yj - yi,  dz = zj - zi
    Возвращает:
        Матрица поворота стержня V
    '''

    # Вспомогательные функции для определения угла между векторами
    dotproduct = lambda v1, v2: sum((a*b) for a, b in zip(v1, v2))
    length = lambda v: math.sqrt(dotproduct(v, v))
    angle = lambda v1, v2: math.acos(dotproduct(v1, v2) / (length(v1) * length(v2)))

    # 1. Определяем угол между проекцией оси стержня на плоскость [YZ] и осью Y
    angYZ_Y = 0
    if dy == 0 and dz == 0:
        pass
   # elif dy == 0 and dx == 0:
        #angYZ_Y = math.pi / 2

    else:
        angYZ_Y = angle([0, 1, 0], [0, dy, dz])
    if dz < 0:
         angYZ_Y = - angYZ_Y

    print(math.degrees(angYZ_Y))

    # 2. Матрица первого поворота вокруг оси X
    V1 = np.array([[1, 0, 0], 
                   [0, math.cos(angYZ_Y), - math.sin(angYZ_Y)], 
                   [0, math.sin(angYZ_Y), math.cos(angYZ_Y)]])

    # 3. Определяем вектор повернутый относительно оси X
    vX = V1.T.dot([dx, dy, dz])
    
    # 4. Определяем угол между вектором vX и осью X
    angvX_X = angle([1, 0, 0], vX)

    # 5. Матрица второго поворота вокруг оси Z
    V2 = np.array([[math.cos(angvX_X), - math.sin(angvX_X), 0], 
                   [math.sin(angvX_X), math.cos(angvX_X), 0], 
                   [0, 0, 1]])

    corAng = angle([dx, dy, 0], [dx, dy, dz])

    # 6. Матрица поворота вокруг собственной оси на угол a
    V3 = np.array([[1, 0, 0], 
                   [0, math.cos(a), - math.sin(a)], 
                   [0, math.sin(a), math.cos(a)]])
    

  
    
    #return V1.dot(V90.dot(V2.dot(V3))).T
    return V1.dot(V2.dot(V3)).T

def getV3(a, dx, dy, dz, deg=False):
    # Вспомогательные функции для определения угла между векторами
    dotproduct = lambda v1, v2: sum((a*b) for a, b in zip(v1, v2))
    length = lambda v: math.sqrt(dotproduct(v, v))
    angle = lambda v1, v2: math.acos(dotproduct(v1, v2) / (length(v1) * length(v2)))
    

    zOz = angle([0, 0, 1], [dx, dy, dz])
    if dz < 0:
        zOz = math.pi / 2 - zOz
    
    print(math.degrees(zOz))


    ang = a
    if deg:
        ang = math.radians(a)
    l = math.sqrt(dx**2 + dy**2 + dz**2)
    cos_a = dx / l
    
    cos_b = math.cos(zOz)
    sin_b = math.sin(zOz)
    sin_a = dz / (l * sin_b)
    cos_a = (1 - sin_a ** 2) ** 0.5

    A = np.array([[cos_a, -sin_a * cos_b], [-sin_a * cos_b, -cos_a]])
    B = np.array([dx / l, dy / l])
    C = np.linalg.solve(A, B)
    print(C)



    cos_g = C[0]
    sin_g = C[1]

    #print(cos_a * cos_g - sin_a * cos_b * sin_g)

    
    V1 = np.array([[cos_a * cos_g - sin_a * cos_b * sin_g, - cos_a * sin_g - sin_a * cos_b * cos_g, sin_a * sin_b], 
                   [sin_a * cos_g + cos_a * cos_b * sin_g, - sin_a * sin_g + cos_a * cos_b * cos_g, - cos_a * sin_b], 
                   [sin_b * sin_g, sin_b * cos_g, cos_b]])
    
    V3 = np.array([[1, 0, 0], 
                   [0, math.cos(ang), - math.sin(ang)], 
                   [0, math.sin(ang), math.cos(ang)]])
    return V1

```

```python
def V4(a, dx, dy, dz, deg=False):
    # Вспомогательные функции для определения угла между векторами
    dotproduct = lambda v1, v2: sum((a*b) for a, b in zip(v1, v2))
    length = lambda v: math.sqrt(dotproduct(v, v))
    cosAngle = lambda v1, v2: dotproduct(v1, v2) / (length(v1) * length(v2))
    sinFromCos = lambda cos: math.sqrt(1 - cos ** 2)
    ang = a
    if deg:
        ang = math.radians(a)

    # Считаем в положительном квадранте, потом зеркалим, если есть отрицательные направления
    
    l = math.sqrt(dx**2 + dy**2 + dz**2)
    
   

    

    cos_x0z = abs(cosAngle([dx, dy, dz], [0, 0, 1]))
    sin_x0z = sinFromCos(cos_x0z)
    
    
    '''
    if dx >= 0 and dz >= 0:
        cos_x0z = -cos_x0z
    elif dx >= 0 and dz < 0:
        cos_x0z = -cos_x0z
        sin_x0z = -sin_x0z
    elif dx < 0 and dz < 0:
        sin_x0z = -sin_x0z
    elif dx >= 0 and dz < 0:
        pass
    '''


    
    cos_t = 0
    if dx != 0 and dy != 0:
        cos_t = abs(cosAngle([dx, dy, 0], [1, 0, 0]))
    elif dx == 0 and dy == 0:
        cos_t = 1
    elif dx != 0 and dy == 0:
        cos_t = 1
    sin_t = sinFromCos(cos_t)
    
    

    #cos_t = abs(cosAngle([rdx, rdy, 0], [1, 0, 0]))
    #sin_t = sinFromCos(cos_t)
    
    if dz > 0:
        if dx >= 0 and dy >= 0:
            cos_t = -cos_t
            sin_t = -sin_t
        elif dx < 0 and dy >= 0:
            sin_t = -sin_t
        elif dx < 0 and dy < 0:
            pass
        elif dx >= 0 and dy < 0:
            cos_t = -cos_t
    else:
        if dx >= 0 and dy >= 0:
            pass
        elif dx < 0 and dy >= 0:
            cos_t = -cos_t
        elif dx < 0 and dy < 0:
            cos_t = -cos_t
            sin_t = -sin_t
        elif dx >= 0 and dy < 0:
            sin_t = -sin_t
    

    print(cos_t, sin_t)
    # Матрицаповорота тержня исходя из положения оси z0 ЛСК в плоскости {x0z} drz всегда положтительно
    V = np.array([[dx / l, dy / l, dz / l], [0, 0, 0], [cos_t * abs(cos_x0z), sin_t * abs(cos_x0z), sin_x0z]])
    V[1] = np.cross(V[2], V[0])

    #V = np.diag([1, -1, 1]).dot(V) # Почему то нужно инвертировать оь У...

   


    
    # Матрица поворота вокруг собственной оси на угол a
    V3 = np.array([[1, 0, 0], 
                   [0, math.cos(ang), - math.sin(ang)], 
                   [0, math.sin(ang), math.cos(ang)]]).T

    return V3.dot(V)
```
